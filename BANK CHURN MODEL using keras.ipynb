```python
import pandas as pd
import matplotlib as plt
df = pd.read_csv("csvs\\churndata.csv")
df
```

```python
df.drop('customerID', axis= 1, inplace = True)
df
```

```python
df['gender'].replace({'Female':1,'Male':0},inplace=True)
```

```python
def print_unique(df):
    for col in df:
        if df[col].dtypes == 'object':
            print(f'{col} : {df[col].unique()}') 
        
```

```python
print_unique(df)
```

```python
pd.to_numeric(df.TotalCharges)
```

```python
df.iloc[488]
```

```python
df1 = df[df['TotalCharges'] != ' ']
df1.TotalCharges = pd.to_numeric(df1.TotalCharges)
```

```python
df1.shape
```

```python
df1.replace('No internet service','No',inplace=True)
df1.replace('No phone service','No',inplace=True)
```

```python
print_unique(df1)
```

```python
df1.replace({'DSL' : 'Yes', 'Fiber optic' : 'Yes'}, inplace = True)
print_unique(df1)
```

```python
df1.replace({'Yes': 1, 'No' : 0}, inplace = True)
print_unique(df1)
```

```python
df2 = pd.get_dummies(data = df1, drop_first = True, dtype = float, columns = ['Contract','PaymentMethod'])
df2
```

```python
cols_to_scale = ['tenure','MonthlyCharges','TotalCharges']

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df2[cols_to_scale] = scaler.fit_transform(df2[cols_to_scale])
```

```python
X = df2.drop('Churn',axis='columns')
y = df2['Churn']

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=5)
```

```python
df2.dtypes
```

```python
X_train.shape
```

```python
import tensorflow as tf
from tensorflow import keras

model = keras.Sequential([
    keras.layers.Dense(20, input_shape = (22,), activation = 'relu'),
    keras.layers.Dense(1, activation = 'sigmoid')
])
model.compile(optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy'])
model.fit(X_train, y_train, epochs = 30)
```

```python
model.evaluate(X_test, y_test)
```

```python
yp = model.predict(X_test)
yp[:5]
```

```python
y_pred = []
for element in yp:
    if element > 0.5:
        y_pred.append(1)
    else:
        y_pred.append(0)
```

```python
from sklearn.metrics import confusion_matrix , classification_report

print(classification_report(y_test,y_pred))
```

```python
import seaborn as sn
%matplotlib inline
cm = tf.math.confusion_matrix(labels=y_test,predictions=y_pred)
plt.figure(figsize = (10,7))
sn.heatmap(cm, annot=True, fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
